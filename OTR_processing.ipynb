```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import json

# File path to save the dictionary
file_path = "/home/luke/SoilAnalysis/element_symbols.json"

# Save the dictionary to a JSON file
with open(file_path, "r") as file:
    element_symbols = json.load(file)
```

```python
df = pd.read_csv('/home/luke/SoilAnalysis/OTR Soil Inorganics 1991 Genaralized Locations for Open Data Portal Oct 2014.csv')
# Remove unnecessary columns
df = df.iloc[:, 4:-3]
# Some columns are in % -> convert to ppm
df = df.apply(lambda col: col * 10**6 if col.name.endswith('%') else col)
df = df.rename(columns=lambda x: x[:-3] if x.endswith('%') else x)
# Rename columns to element names
df = df.rename(columns=lambda x: element_symbols.get(str.lower(x.split('_')[0]), x))
#df.rename(columns=lambda x: element_symbols[str.lower(x)])
```
